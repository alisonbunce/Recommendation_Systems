## 2. Understand Short Description Column Data

#### First let's extract the column we want, make sure they are free of null values

```python
#create new df for short descriptions dropping all null values
Sdescription = data[['Products.short_description']].dropna()
```

```python
#double check that all null values are dropped
Sdescription.isnull().sum()
```

## 3. Use TFIDF and turn these words into vectors

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
word_vec = vectorizer.fit_transform(Sdescription['Products.short_description'])
len(vectorizer.vocabulary_)
```

### 3.1 Observe our libarary of vectors and their corresponding numerical meaning

```python
# exhibits the index of each word
vectorizer.vocabulary_
```

### 3.2 Show us the row, column with corresponding TFIDF ratio

```python
# row, column, Tfidf ratio 
#Tfidf ratio = % of each word in the setence / % of each word in the document 
print(word_vec)
```

```python
print(word_vec)
```

### 4. Create a Dataframe using CoSine Similarity based on our TFIDF metrics (above)

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
from scipy.sparse import csr_matrix
from scipy.sparse.linalg import svds
```

```python
pd.DataFrame(cosine_similarity(word_vec)).head()
```

## 5. Use Count Vectorizer 

```python
# vectorizer = CountVectorizer()
description_vector = CountVectorizer().fit_transform(Sdescription['Products.short_description'].values)
```

```python
print(description_vector)
```

```python
#now we have take the cosine similarity of the description vector
print(cosine_similarity(description_vector))
```
