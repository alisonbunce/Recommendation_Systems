## 1.2 Import Data

## 2. Analyze the Data

### 2.1 Create a Pivot Table

```python
#create a pivot table, it shows us specifically the specifc data we want to compare
df = data.pivot_table(index=['Customers.id'], columns=['Products.name'], values='Order_Items.qty')
df.head()
```

## 3. Clean Data

```python
#replace the null values with 0
df.fillna(0, inplace=True)
```

```python
#double check that its done successfully
df.head()
```

analysis note: this is what we will use for our utility matrix

```python
#double check all values still there, looks good
df.shape
```

## 4. Convert our pivot table into a matrix

utility matrix now has everything filled in

now...de-mean the data (normalize by each users mean) and convert it from a dataframe to a numpy array

```python
#convert the pivot table, just in a df now, into a matrix
print(df.shape)
R = df.as_matrix()
R
```

now we have a matrix we can use. lets calculate the mean for each user, and subtract the mean from each order/row (de mean the data)

## 5. Demean the Data

```python
#demean the data, and show it to us for the first 5 users
import numpy as np
df_mean = np.mean(R, axis = 1)
df_mean[:5]
```

```python
#no columns here, just rows 
df_mean.shape
```

```python
R.shape
```

```python
#now we subtract the mean from each element of the row
```

```python
#remove the user matrix mean from 
# -1 keep the rows, how every many you have
# 1 i want 1 column, give me 1 column
```

```python
#we are creating a column for each row (note in df_mean.shape shows no columns). the new column is for the mean
df_mean.reshape(-1, 1).shape
```

```python
# take the matrix R and subtract the mean (the actual de-meaning)
R_demeaned = R - df_mean.reshape(-1,1)
R_demeaned
```

```python
#double check that we have the same # of rows and columns as we did in the orig matrix R, but these are all demeaned
R_demeaned.shape
```

## 6. Singular Value Decompostion

Use SVD to convert the one big matrix R_demeaned, into 3 smaller matrixes

```python
from scipy.sparse.linalg import svds

#u is the customer id  matrix, which is all the users
# R is the utility matrix
# v transpose is the product matrix (can be whatver you want ot call it)
# sigmna is diagonal matrix, is a sqaure matrix that only has values diagonal (has all 0s aside from the values that fall diagonally in the matrix) 
#ask zia about diag matrix***

U, sigma, Vt = svds(R_demeaned, k = 200)
#take those 1391 columns and compress them into 20 (changed to 200)
```

```python
#this is just showing us what we did about (not required)
svds(R_demeaned, k=200)
```

recall that we now have 3 matrixes, 
1. U
2. Sigma
3. Vt

Double check our new matrixes, note that this breaks down every product

```python
#len is always rows
#shape will give you the columns and rows
print(U.shape)
```

```python
#check to see the matrix
U
```

```python
#when you do svd, sigma comes back as an array,not a matrix, you have to convert it 
print(sigma.shape)
sigma
```

```python
#this breaks out every single product
print(Vt.shape)
Vt
```

```python
#convert array output into diagonal matrix
sigma = np.diag(sigma)
sigma
```

```python
#we took the sigma, just 20 columns, and converted it into diag matrix successfully 20x20, this confirms that
sigma.shape
```

#we changed the nulls, converted into into 3 matries, now we have 3 matrizes, now take these 3 u,daig sign, 
do a dot product, of these 3 matrixes to come up with a predicted matrix.

through the use of that is linear algebra, we will be multiply these togther to give us a predicted utility matrix. this will have predicted product quantities


attributing diagnoal values. same number of rows and columns. when you get the result, multiply it by v transpose, not multiplication, its matrix multiplication

```python
#np.dop is the function for matix multiplication
predicted_products = np.dot(np.dot(U, sigma), Vt)
predicted_products
```

Now...based on all these predicted ratings, we convert this into a dataframe. bring all the columns from the pivot table and assign them as columns

## 7. Create a new Dataframe with our predictions

```python
new_df = pd.DataFrame(predicted_products, columns=df.columns)
new_df.head(2)
```

These are our predictions! 

```python
#first users prediction for every product (foucsing on first user)

#we are matching up the values vs predicted values for one particular user
new_df.loc[0].sort_values(ascending=False)
#the algo svd tells us that based on other users and other products, this is most likely what user will buy next. 
#amaz would rec these 3 items first becuase you bouht them last time... based on you as a user
```
